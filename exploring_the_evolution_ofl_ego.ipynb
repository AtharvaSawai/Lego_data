## IMPORTING THE DATASETS AND LIBRARIES REQUIRED

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the DataSet of Lego.
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')

theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')

theme.head()
```

```python
df.head()
```

## TASK #1:What percentage of all licensed sets were ever Released we STAR WARS Themed?

#### 1. Here I can get the count of the sets from the df dataframe.
#### 2. Also we to know whether the set is licensed, that information is give the theme dataframe. So, we would eventually need to join/concatenate/merge (i.e. they all are different in some order, with some hands on experience you eventually figure it out what is required when!!!) the dataframe.
#### 3. After getting the final table we can filter out the count of the licensed sets which are of the 'STAR WAR' theme and then divide it by total count of the sets produced in LEGO.

```python
merged = df.merge(theme, left_on = 'parent_theme', right_on = 'name')

merged.head()
```

```python
# We have the merged dataset, we can perform pre requisites for working with the dataset. 
# Here I observe there is slight change in the column name  'name' -> 'name_x', 
# this smart move is done by pandas for diffrentiating between the columns.

# I want to drop the unnecessary column 'name_y'

merged = merged.drop(columns = 'name_y')
```

```python
merged.head()
```

```python
licensed = merged[merged['is_licensed'] == True]
licensed = licensed.dropna(subset = ['set_num'])
# In here even if we did not write == True it is by default considered as True but so is not the case for vice versa.
```

```python
licensed.head()
```

```python
licensed.shape[0]
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']



star_wars.head()
```

```python
star_wars.shape[0]
```

```python
#to get the percentage we divide the count of all Licensed with Star Wars Licensed.

result1 = int(star_wars.shape[0]/licensed.shape[0] * 100)

print(result1)
```
